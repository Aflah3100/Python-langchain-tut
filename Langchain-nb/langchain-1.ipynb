```python
# Creating agent

import os
from dotenv import load_dotenv
from langchain_groq import ChatGroq
from langchain.agents import create_agent
from langchain.tools import tool


# Load environment variables
load_dotenv()

# Read API key
groq_api_key = os.getenv("GROQ-API-KEY")


model = ChatGroq(
    api_key=groq_api_key,
    model="llama-3.1-8b-instant"
)

@tool('get_user_name',description="Return the user name based on user id",return_direct=True)
def get_user_name(id:str )-> str:
    """ Get the user name based on the user id"""
    match(id):
        case "1":
            return "User name is Aflah"
        case "2":
            return "User name is Ahnaf"
        case _:
            return "User name not found"

@tool('get_user_age',description="Return the age of the user based on the user id")
def get_user_age(id: str)->str:
    match(id):
        case "1":
            return "25"
        case "2":
            return "27"
        case _:
            return "User id doesn't exist"

agent=create_agent(
    model=model,
    tools=[get_user_name,get_user_age],
    system_prompt='You are a helpful assistant'
)


response=agent.invoke({
    "messages":[
        
        {
            "role":"user",  
            "content":"May i know the age of the user with user id 1"

        }
    ]
})

print(response["messages"][-1].content)
```
